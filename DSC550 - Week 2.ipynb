```python
# import pandas library
import pandas as pd
import numpy as np
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# import controversial-comments file
d_frame= pd.read_json('C:/Users/Saurabh/Desktop/controversial-comments.jsonl', lines = True)
d_frame.shape
```

```python
# Converted all text to lowercase letters.

d_frame['txt'].str.lower()
d_frame
```

```python
# Remove all punctuation from the text.

import string
d_frame['txt'] = d_frame['txt'].str.translate(str.maketrans('','',string.punctuation))
d_frame
```

```python
# removed stop words

stop_words=stopwords.words('english')

d_frame['txt'] = d_frame['txt'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
d_frame
```

```python

# creating stemmer
porter = PorterStemmer()

# applying stemmer
d_frame['txt'] = d_frame['txt'].apply(lambda x: ' '.join([porter.stem(word) for word in x.split()]))
d_frame
```
